## Example Science Cases

* **Finding z=7 Quasars**: How many z=7 quasars will we find after 1 year of surveying? How well can these objects be characterised? i.e. How pure a sample can we make? z=7 quasars are recognisable by 1) their z-dropout nature (ie zero flux in _ugriz_ and some detection in _y_), 2) their quasar-like variability (ie lightcurves with variability on all scales, and a increasing structure function), and 3) their lack of proper motion.
* **Finding RR Lyrae**: Likewise, but for RR Lyrae stars, for mapping MW halo structure. RR Lyrae are recognisable by their less-than-one day, highly periodic and in some cases modulated variability, in all bands.
* **Finding Young Stellar Object (YSO) Bursts**: How well can we characterise the population of young stellar object bursts? Young stars undergo possibly episodic bursts, where their brightness increases rapidly by ~1 mag, before decaying over several weeks. _u_-band variability is particularly interesting. Distinguishing YSO bursts from other transients relies on the presence of a YSO precursor. 
* **Measuring Lensed Quasar Time Delays**: How many lensed quasars will have well-measured time delays? The requirements for accurate cosmology are extreme, but the TDC model allows us to relate the accuracy to three summaries of the observing pattern, night-to-night cadence, season length and campaign length.

## Common Ground

Many science projects involve _using LSST as a finder_. In these cases, the main science comes from spectroscopic, multi-wavelength and/or high resolution imaging follow-up of an interesting sample of objects. In these projects, there will be keen interest in following up these new targets very soon after the first data release. Similarly, it is of interest to know how the sample is likely to evolve after 2, 3, years. _We would like to investigate metrics relating to these science cases over the course of the campaign._

The metrics of interest to each group relate to either the detectability or measurability of time-variable objects. As such they are likely to be functions of the `night` or `expMJD` database fields. In each case we expect to have a model of how the objects across the sky, and so knowing the detectability/measurability across the sky should be useful in science planning - so we expect the `healpixSlicer` (with `ditheredRA` and `ditheredDec`) as being most relevant.

## Possible Tensions

RR Lyrae potentially benefit from intra-night sampling (ie the solar system visit pairs), while longer timescale variable AGN might benefit from spreading out the visits more. 

YSOs and high extinction RR Lyrae need observations in the galactic plane, while quasar searching and measurement might be better focused on the extragalactic sky.

## Periodic Variable Metric Analysis

The Sims team has started developing some variability metrics to begin understanding LSST performance for discovering and characterising periodic variables. This notebook has been adapted from [Variability Metrics.ipynb](https://github.com/LSST-nonproject/sims_maf_contrib/blob/master/other/Variability%20Metrics.ipynb), and uses the metrics in [varMetrics.py](https://github.com/LSST-nonproject/sims_maf_contrib/blob/master/mafContrib/varMetrics.py), all in the [sims_maf_contrib](https://github.com/LSST-nonproject/sims_maf_contrib) github repo.

#### Periodic Variable Metrics 

One metric, `PhaseGaps`, looks at the gaps in phase coverage for a set of observations. At each point in the sky (if we use a healpix slicer), a range of periods and the times of the visits are converted to phase values: then the largest gap in these phases is calculated. The metric calculates the max gap for a range of periods at each point; thus it also provides a few reduce functions to convert this to the period with the largest gap (at each point in the sky) and the value of the largest gap in phase coverage (at each point in the sky). We can also read the metric values back into python and generate plots of phase gaps vs. period, including the data for all points.

A second metric. `PeriodDeviation`, generates a light curve (representing the variability of the source as a simple sine curve but neglecting photometric errors) with user-settable range of periods, then attempts to recover the period using a simple lomb-scargle calculation. Changing the period fitting routine would doubtless improve the period recovery.  Again, this should be paired with a healpix slicer to evaluate sets of visits which correspond to the same spot on the sky, and again each point is sampled at a range of different periods so the metric provides reduce methods and we can also read back in the results to evaluate true period vs. fit period across the entire sky.

Set up the metrics and slicer. 

```python
# Metric to calculate the time between first and last observation

class LightCurveLengthMetric(metrics.BaseMetric):
    """
    Calculate the time between first and last observation
    """
    def __init__(self, tCol='expMJD', **kwargs):
        """
        tCol = the name of the column defining the visit MJD
        """
        self.tCol = tCol
        super(LightCurveLengthMetric, self).__init__(col=[self.tCol], **kwargs)

    def run(self, dataSlice, slicePoint=None):
        lclength = np.max(dataSlice[self.tCol]) - np.min(dataSlice[self.tCol])
        return lclength
```

```python
# phaseMetric = mafContrib.PhaseGapMetric(nPeriods=20, periodMin=0.2, periodMax=3.5)
# periodMetric = mafContrib.PeriodDeviationMetric(nPeriods=2, periodMin=2, periodMax=3.5)
# phaseslicer = slicers.HealpixSlicer(nside=32, lonCol='ditheredRA', latCol='ditheredDec')
# periodslicer = slicers.HealpixSlicer(nside=4, lonCol='ditheredRA', latCol='ditheredDec')
lclengthMetric = LightCurveLengthMetric()
lclengthSlicer = slicers.HealpixSlicer(nside=32, lonCol='ditheredRA', latCol='ditheredDec')
```

```python
summaryMetrics = [metrics.MinMetric(), metrics.MaxMetric(), metrics.MeanMetric()]
```

```python
def process(runName, startNight = 1, endNight = 365):
    """
    method to process a run, queryin the database, calculating metrics and make derived plots
    runName: name of run to be processed
    """
    
    dbAddress = 'sqlite:///' + runName + '_sqlite.db'
    opsdb = db.OpsimDatabase(dbAddress)
    outDir = runName + "_outputs" 

    lclengthsqlconstraint = 'night>=' + str(startNight) + ' and night<= ' + str(endNight) + ' and filter="y"'
    # phaseBundle = MetricBundle(PhaseMetric, phaseslicer, sqlconstraint=sqlconstraint, 
    #                            runName=runName, summaryMetrics=summaryMetrics)
    # periodBundle = MetricBundle(periodMetric, periodslicer, sqlconstraint=sqlconstraint, 
    #                             runName=runName, summaryMetrics=summaryMetrics)
    lclengthBundle = MetricBundle(lclengthMetric, lclengthSlicer, sqlconstraint=lclengthsqlconstraint, 
                            runName=runName, summaryMetrics=summaryMetrics)
 
    bdict = {'LCLength':lclengthBundle}

    resultsDb = db.ResultsDb(outDir=outDir)  
    bgroup = MetricBundleGroup(bdict, opsdb, outDir=outDir, resultsDb=resultsDb)

    bgroup.runAll()

    bgroup.plotAll(closefigs = False)
```
